Let $R$ be predicate symbol of arity $1$. Show that $\exists x (R(x) \Rightarrow \forall y R(y))$ is logically valid.
    
#### Solution

Suppose $\exists y \neg R(y)$. 

So we can pick $a$ such that $\neg R(a)$.

This way he have that $\exists y \neg R(y) \Rightarrow \neg R(a)$.

So, $\exists x (\exists y \neg R(y) \Rightarrow \neg R(x))$ is logically valid.

But we know that $\exists x (\exists y \neg R(y) \Rightarrow \neg R(x)) \equiv \exists x (R(x) \Rightarrow \forall R(y))$.

So, $\exists x (R(x) \Rightarrow \forall R(y))$ is also logically valid.

But we supposed $\exists y \neg R(y)$. So now we must suppose the oposite.

Suppose $\forall y R(y)$.

In this case $R(x) \Rightarrow \forall y R(y)$ is logically valid for any $x$ because the implication is aways true.

So we have $\forall x (R(x) \Rightarrow \forall y R(y))$ logically valid and this implies that $\exists x (R(x) \Rightarrow \forall y R(y))$ is also logically valid.

### Exercise 2

Show that the following sentence is unsatisfiable, where $S$ is any formula with two free variables: $ \exists x \forall y (S(x,y) \Leftrightarrow \neg S(y,y))$.

#### Solution

Pick $x = a$ that makes the sentence satisfiable, so $\forall y (S(a,y) \Leftrightarrow \neg S(y,y))$.

For $y = a$ we have $S(a,a) \Leftrightarrow \neg S(a,a)$, which is a contradiction. 

So, $\exists x \forall y (S(x,y) \Leftrightarrow \neg S(y,y))$ is unsatisfiable.

### Definition

$\mathfrak{B}$ is a substructure of $\mathfrak{A}$ if: $B \subseteq A$; for every constant symbol $c$, $c^\mathfrak{A} = c^\mathfrak{B}$, every relation $R^\mathfrak{B}$ (resp. function $f^\mathfrak{B}$) is the restriction of $R^\mathfrak{A}$ (resp. function $f^\mathfrak{A}$) to $B$.

### Exercise 1

**(i)** If $\mathfrak{B}$ is a substructure of $\mathfrak{A}$, then for any atomic formula $F(x_1, \cdots, x_n)$ for all $b_1, \cdots, b_n$ in $B$, $\mathfrak{B} \models F[b_1, \cdots, b_n]$ iff $\mathfrak{A} \models F[b_1, \cdots, b_n]$. **(ii)** Let $T$ be a set of parely universal sentences (i.e. sentences starting with universal quantifiers followed by an atomic formula). If $\mathfrak{B}$ is a substructure of $\mathfrak{A}$ and $\mathfrak{A} \models T$ then $\mathfrak{B} \models T$.

#### Solution

**(i)**

An atomic formula, by the definition, is a relation $R(t_1, \cdots, t_n)$ or a string of the type $(t = s)$.

So, proving the following 

$\mathfrak{B} \models F[b_1, \cdots, b_n] \Leftrightarrow \mathfrak{A} \models F[b_1, \cdots, b_n]$, for any atomic formula $F(x_1, \cdots, x_n)$ for all $b_1, \cdots, b_n$ in $B$

is the same of proving both:

$S_1$: $\mathfrak{B} \models R[b_1, \cdots, b_n] \Leftrightarrow \mathfrak{A} \models R[b_1, \cdots, b_n]$ for any relation $R(t_1, \cdots, t_n)$ for all $b_1, \cdots, b_n$ in $B$

and

$S_2$: $\mathfrak{B} \models (b_1 = b_2) \Leftrightarrow \mathfrak{A} \models (b_1 = b_2)$ for all $b_1, b_2 \in B$.

Lets first prove $S_1$:

$\mathfrak{B} \models R[b_1, \cdots, b_n]$

$\Leftrightarrow$

$(b_1, \cdots, b_n) \in R^\mathfrak{A}$ by the definition of satisfaction

$\Leftrightarrow$

$(b_1, \cdots, b_n) \in R^\mathfrak{B}$ because $(b_1, \cdots, b_n) \in B \wedge B \subseteq A \Rightarrow (b_1, \cdots, b_n) \in A$ and $R^\mathfrak{B}$ is a restriction of $R^\mathfrak{A}$ in $B$

$\Leftrightarrow$

$\mathfrak{A} \models R[b_1, \cdots, b_n]$ by the definition of satisfaction.

Now, lets prove $S_2$ which is is trivial:

$\mathfrak{B} \models (b_1 = b_2)$ 

$\Leftrightarrow$ 

$\mathfrak{A} \models (b_1 = b_2)$ because $b_1, b_2 \in B \wedge B \subseteq A \Rightarrow b_1, b_2 \in A$.

**(ii)**

$T$ is a set of sentences of the type $S = \forall x_1 \cdots \forall x_k F(x_1, \cdots, x_k)$, where $F$ is a an atomic formula).

We have that

$\mathfrak{A} \models T$

$\Leftrightarrow$

$\mathfrak{A} \models S$ for any $S$ in $T$

$\Leftrightarrow$

$\mathfrak{A} \models \forall x_1 \cdots \forall x_n F(x_1, \cdots, x_n)[\alpha]$

$\Leftrightarrow$

$\forall a_1 \cdots \forall a_n \in A$, $\mathfrak{A} \models F\left[\alpha\begin{pmatrix}
x_1 \cdots x_n \\ a_1 \cdots a_n
\end{pmatrix}\right]$ by the definition of satisfaction.

Similarly, we have that

$\mathfrak{B} \models T$

$\Leftrightarrow$

$\forall b_1 \cdots \forall b_n \in B$, $\mathfrak{B} \models F\left[\beta\begin{pmatrix}
x_1 \cdots x_n \\ b_1 \cdots b_n
\end{pmatrix}\right]$.

So, we have that

$\mathfrak{A} \models T$

$\Leftrightarrow$

$\forall a_1 \cdots \forall a_n \in A$, $\mathfrak{A} \models F\left[\alpha\begin{pmatrix}
x_1 \cdots x_n \\ a_1 \cdots a_n
\end{pmatrix}\right]$

$\Rightarrow$

$\forall b_1 \cdots \forall b_n \in B$, $\mathfrak{A} \models F\left[\beta\begin{pmatrix}
x_1 \cdots x_n \\ b_1 \cdots b_n
\end{pmatrix}\right]$ because $B \subseteq A$

$\Leftrightarrow$

$\forall b_1 \cdots \forall b_n \in B$, $\mathfrak{B} \models F\left[\beta\begin{pmatrix}
x_1 \cdots x_n \\ b_1 \cdots b_n
\end{pmatrix}\right]$ from (i)

$\Leftrightarrow$

$\mathfrak{B} \models T$.

So, finally, $\mathfrak{A} \models T \Rightarrow \mathfrak{B} \models T$.

### Definition

$\mathfrak{B}$ substructure of $\mathfrak{A}$ is called elementary if for all formulas $F(x_1, \cdots, x_n)$ for all $b_1, \cdots, b_n$ in $B$, $\mathfrak{A} \models F[b_1, \cdots, b_n]$ iff $\mathfrak{B} \models F[b_1, \cdots, b_n]$. ### Exercise 2

Prove the following: $\mathfrak{B}$ a substructure of $\mathfrak{A}$ is elementary iff for every formula $F(x_1, \cdots, x_n)$ and for every $b_1, \cdots, b_n$ in $B$: If for some $a \in A$ we have $\mathfrak{A} \models F[b_1, \cdots, b_n, a]$ then there exists $b \in B$ such that $\mathfrak{B} \models F[b_1, \cdots, b_n, b]$.

#### Solution

For $\mathfrak{B}$ a substructure of $\mathfrak{A}$ and $b_1, \cdots, b_n \in B$, and for

$S_1$: $\mathfrak{A}$ and $\mathfrak{B}$ agree on elements of $B$

$S_2$: $\exists a \in A, \mathfrak{A} \models F[b_1,\cdots,b_n,a] \Rightarrow \exists b \in B, \mathfrak{B} \models F[b_1,\cdots,b_n,b]$

we want to prove that $S_1 \Leftrightarrow S_2$.

We can rewrite $S_2$ as

$\forall b \in B, \mathfrak{B} \not\models F[b_1,\cdots,b_n,b] \Rightarrow \forall a \in A, \mathfrak{A} \not\models F[b_1,\cdots,b_n,a]$.

But, $B \subseteq A$, so we can rewrite $S_2$ again as

$\forall b \in B, \mathfrak{B} \not\models F[b_1,\cdots,b_n,b] \Rightarrow \forall b \in B, \mathfrak{A} \not\models F[b_1,\cdots,b_n,b]$.

Simplifying we have that

$\forall b \in B, (\mathfrak{B} \not\models F[b_1,\cdots,b_n,b] \Rightarrow \mathfrak{A} \not\models F[b_1,\cdots,b_n,b])$. ●

Now, suppose $S_2$ is not true (using $S_2$ writed of the form ●), so we have

$\exists b \in B, (\mathfrak{A} \models F[b_1,\cdots,b_n,b] \wedge \mathfrak{B} \not\models F[b_1,\cdots,b_n,b])$.

Which implies that $\mathfrak{A}$ and $\mathfrak{B}$ do not agree on elements of $B$.

At this point we proved that $\neg S_2 \Rightarrow \neg S_1$, which is equivalent that $S_1 \Rightarrow S_2$.

Now, we have to prove the other direction.

Supose $S_2$ is true (using $S_2$ writed of the form ●), so we have the following for all formula $F$:

$\forall b \in B, (\mathfrak{B} \not\models F[b_1,\cdots,b_n,b] \Rightarrow \mathfrak{A} \not\models F[b_1,\cdots,b_n,b])$. ●

Make $G$ be $\neg F$, so, by the definition of satisfaction, we also have the following for all formula $G$:

$\forall b \in B, (\mathfrak{B} \models G[b_1,\cdots,b_n,b] \Rightarrow \mathfrak{A} \models G[b_1,\cdots,b_n,b])$. ●●

● and ●● implies that $\mathfrak{A}$ and $\mathfrak{B}$ agree on elements of $B$.

By now, we also proved that $S_2 \Rightarrow S_1$, which is the other direction

So we proved $S_1 \Leftrightarrow S_2$.

### Exercise 3

Show that the structure $Q = (\mathbb{Q}, +, \times, 0, 1)$ is a substructure of structure $R= (\mathbb{R}, +, \times, 0, 1)$, but not elementary.

#### Solution

$Q$ is a substructure of $R$ if:

- $\mathbb{Q} \subseteq \mathbb{R}$ - TRUE, because all rational numbers are also real numbers.
- For every constant symbol $c$, $c^R = c^Q$ - TRUE, because $0 = 0$ and $1 = 1$.
- Every relation $E^Q$ is the restriction of $E^R to \mathbb{Q}$ - TRUE, because $\mathbb{Q} + \mathbb{Q} \rightarrow \mathbb{Q}$ and $\mathbb{Q} \times \mathbb{Q} \rightarrow \mathbb{Q}$.

So, $Q$ is a substructure of $R$.

$Q$ substructure of $R$ is called elementary if for all formulas $F(x_1, \cdots, x_n)$ for all $q_1, \cdots, q_n$ in $B$, $R \models F[q_1, \cdots, q_n]$ iff $Q \models F[q_1, \cdots, q_n]$. This is FALSE, an example that makes it fails could be $F(x) = \exists y$, $y \times y = x$.

In this example we have that

$\forall q \in \mathbb{Q}, R \models \exists y$, $y \times y = q$, because $y = \sqrt{q}$ and $\sqrt{q} \in \mathbb{R}$,

and

$\forall q \in \mathbb{Q}, Q \not\models \exists y$, $y \times y = q$, because, for example, for $q = 2$, there is no $y \in \mathbb{Q}$ that satisfies the formula.

So $R$ and $Q$ does not agree on elements of $\mathbb{Q}$.
